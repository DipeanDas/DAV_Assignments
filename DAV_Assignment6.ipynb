# LAB TASK

**TASK1:**<br>
**Create a column named “average_rate” in “primary_completion_rate_female.csv” that
contains the average (over years) completion rate for each country.**

```python
# Defining a custom function to fill null values with row means for numeric columns only
def fill_null_with_row_mean(row):
    num_val = pd.to_numeric(row, errors='coerce')
    if num_val.notnull().any():
        r_mean = num_val.mean()
        row.fillna(r_mean, inplace=True)  #fill the null places with mean of that row
    return row

pcrf = pcrf.apply(fill_null_with_row_mean, axis=1)
```

```python
pcrf['average_rate'] = pcrf.iloc[:, 4:].mean(axis=1, skipna=True)
```

```python
pcrf.info()
```

```python
pcrf1=pcrf[['Country Name', 'average_rate']]
pcrf1
```

**TASK3<br>
What is the weighted average of “average_rate”? The weight should be calculated based on “latest population” of females in the country.**

```python
lt_pop = pof[['Country Name', '2021']]
df_merge = pd.merge(pcrf, lt_pop, on='Country Name') #merging pcrf and lt_pop on Country Name field
```

```python
df_merge['Weighted_Average'] = df_merge['average_rate'] * df_merge['2021_y']
```

```python
tot_weighted_sum = df_merge['Weighted_Average'].sum()
tot_pop = df_merge['2021_y'].sum()
```

```python
weighted_avg = tot_weighted_sum / tot_pop
```

```python
print("Weighted Average of 'average_rate':", weighted_avg)
```

**TASK4<br>
What is the Median of “average_rate”?**

```python
median_avg_rate = df_merge['average_rate'].median()
print("Median of 'average_rate':", median_avg_rate)
```

**TASK5<br>
What is the InterQuartile Range of “average_rate”?**

```python
# Calculating the 25th and 75th percentiles
Q1 = df_merge['average_rate'].quantile(0.25)
Q3 = df_merge['average_rate'].quantile(0.75)

# Calculating the Interquartile Range
IQR = Q3 - Q1

print("InterQuartile Range of 'average_rate':", IQR)
```

**TASK6<br>
 What is the Variance of “average_rate”?**

```python
var_avg_rate = df_merge['average_rate'].var()

print("Variance of 'average_rate':", var_avg_rate)
```
