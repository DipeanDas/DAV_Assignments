**TASK**

```python
ds= pd.DataFrame(DS)
```

```python
plt.figure(figsize=(12, 6))
sns.pointplot(data=ds[['Year', 'Food Grains (Cereals) - Rice (000 tonnes)']], x='Year', y='Food Grains (Cereals) - Rice (000 tonnes)')
plt.title('Yearly Rice Production')
plt.xlabel('Year')
plt.ylabel('Rice Production (000 tonnes)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 10))

# Creating the boxplot
sns.boxplot(data=ds[['Banana (000 tonnes)','Potatoes (000 tonnes)','Coconut (000 tonnes)']])
plt.title('Banana Potato and Coconut Production')
plt.ylabel('Production (000 tonnes)')
plt.xticks(rotation=45)

# Calculate the maximum value for all cereal columns
max_value = ds[['Banana (000 tonnes)','Potatoes (000 tonnes)','Coconut (000 tonnes)']].max().max()
plt.yticks(np.arange(0, max_value + 1, 1000))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=ds[['Food Grains (Cereals) - Maize (000 tonnes)', 'Food Grains (Cereals) - Jowar (000 tonnes)','Food Grains (Cereals) - Bajra (000 tonnes)']])
plt.title('Maize Jowar and Bajra Production')
plt.ylabel('Production (000 tonnes)')
plt.xticks(ticks=[0, 1, 2], labels=['Maize', 'Jowar','Bajra'])
plt.show()
```

```python
plt.figure(figsize=(14, 10))
joyplot(data=ds[['Food Grains (Pulses) - Gram (000 tonnes)', 'Food Grains (Pulses) - Tur (000 tonnes)','Food Grains (Pulses) - Other Pulses (000 tonnes)']], overlap=0)
plt.title('Pulses Distribution')
plt.xlabel('Production (000 tonnes)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

total_production_columns = ['Food Grains (Cereals) - Total (000 tonnes)',
                            'Food Grains (Pulses) - Total (000 tonnes)',
                            'Total Nine Oilseeds (000 tonnes)']

total_production_yearly = ds.groupby('Year')[total_production_columns].sum()

total_production_yearly.plot(kind='bar', width=0.7)

plt.title('Total Production of Food Grain Cereals, Pulses, and Total Nine Oilseeds (Year-wise)')
plt.xlabel('Year')
plt.ylabel('Total Production (000 tonnes)')

plt.show()

```

```python
plt.figure(figsize=(12, 8))

# Creating the boxplot
sns.boxplot(data=ds[['Jute (000 Bales)','Cotton (000 Bales)']])
plt.title('Jute and Cotton Production')
plt.ylabel('Production (000 tonnes)')
plt.xticks(rotation=45)

# Calculate the maximum value for all cereal columns
max_value = ds[['Jute (000 Bales)','Cotton (000 Bales)']].max().max()
plt.yticks(np.arange(0, max_value + 1, 1000))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=ds[['Tea (Million kgs)','Coffee (000 MT)']])
plt.title('Tea and Coffee Production')
plt.ylabel('Production (000 Mtonnes)')
plt.xticks(ticks=[0, 1], labels=['Tea', 'Coffee'])
plt.show()
```
