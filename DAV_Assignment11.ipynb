# **LAB TASK**

**TASK1:Perform t-test on cookie cats data and make conclusions from derived p-value.**

```python
vA_data = DF1[DF1['version'] == 'gate_30']['sum_gamerounds']
vB_data = DF1[DF1['version'] == 'gate_40']['sum_gamerounds']

# Performing t-test
res = stats.ttest_ind(vA_data, vB_data, equal_var=False)

# Printing t-value and degrees of freedom
print(f"t-value: {res.statistic:.4f}")
print(f"p-value for single-sided test: {res.pvalue / 2:.4f}")
print(f"p-value:{res.pvalue:.4f}")

# Make conclusions from derived p-value
alpha = 0.05

if res.pvalue / 2 < alpha:
    print("Reject the null hypothesis. There is a significant difference between the two versions.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the two versions.")

```

**TASK2:Create contingency table from ”Specialism” and ”Batting” type of player from IPL Data. Perform Chi square test on data in contingency table.**

```python
# Taking'Specialism' and 'Batting' as the relevant columns
contgen_t = pd.crosstab(DF2['Specialism'], DF2['Batting'])

# Displaying the contingency table
print('TABLE_of_CONTINGENCY')

# Performing the Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contgen_t)

# Printing the results of the Chi-square test
print("\nChi-square Test Result :")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")

# Make conclusions from derived p-value
alpha = 0.05

if p_value < alpha:
    print("\nReject the null hypothesis. There is a significant association between Specialism and Batting.")
else:
    print("\nFail to reject the null hypothesis. There is no significant association between Specialism and Batting.")
```
